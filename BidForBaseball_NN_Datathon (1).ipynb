# Next Season Predictions for the 150 Free Agency Players



```python
# Read the submission DataFrame
submission = pd.read_csv("submission.csv")  # Adjust the file path as per your requirement

# Read the average player stats DataFrame
avg_df = pd.read_csv("average_player_stats.csv")  # Adjust the file path as per your requirement

# Merge the two DataFrames on the "Player" column
freeAgency_df = pd.merge(submission, avg_df, on="Name", how="inner")

# Save the new DataFrame to a CSV file
freeAgency_df.to_csv("freeAgency_df.csv", index=False)  # Adjust the file path as per your requirement

# Display the new DataFrame
freeAgency_df
```

# Create, Format, and List Predictions in Submission File

```python
#That looks great. But now I want to build a forecast.

# Assuming data is already loaded into a pandas DataFrame with the name 'data'
freeAgency_df.sort_values(by=['Name', 'Year'], inplace=True)

# Define the alpha value for EWMA calculation
alpha_value = 0.1  # Replace this with the desired alpha value

# Group by player name and calculate EWMA for hits
freeAgency_df['ewma_hits'] = freeAgency_df.groupby('Name')['H'].transform(lambda x: x.ewm(alpha=alpha_value).mean())

# Select the last EWMA value as the next year's forecast
freeAgency_df['Expected Number of Hits Next Season'] = freeAgency_df.groupby('Name')['ewma_hits'].transform('last')

# Now create the 'submission' DataFrame
submission = freeAgency_df[['Name', 'Expected Number of Hits Next Season']].drop_duplicates()

submission
# freeAgency_df.columns
```

# Bidding Algorithm



Get submission ready for bid values

```python
# Now lets get only the players we need to forecast. We'll need to load in the example submission to do this
player_submission_table = pd.read_csv('/content/drive/MyDrive/example_data/batting_season_summary.csv')
name_list = player_submission_table['Name']
submission = submission[submission['Name'].isin(name_list)]
```

Sort submission, and bid equally on the bottom 10 of the top 30 hitters

```python
# DONT ADD A $ IN YOUR BID COLUMN, IT SHOULD BE A FLOAT OR INT
submission_sorted = submission.sort_values(by='Expected Number of Hits Next Season', ascending=False)


# Select the top 30 players with the highest expected number of hits
top_30_players = submission_sorted.head(30)

# Select the bottom 10 players from the top 30
bottom_10_players = top_30_players.tail(10)

# Calculate the bid amount per player based on the budget and the number of players
bid_amount_per_player = 200 / len(bottom_10_players)

# Set the bid amount for all players to 0 initially
submission['Bid Amount($)'] = 0

# Update the bid amount for the bottom 10 players
submission.loc[bottom_10_players.index, 'Bid Amount($)'] = bid_amount_per_player

# Ensure the DataFrame has the correct columns
submission = submission[['Name', 'Expected Number of Hits Next Season', 'Bid Amount($)']]

# Update the 'submission' DataFrame with the modified bid amounts
submission.loc[bottom_10_players.index, 'Bid Amount($)'] = bid_amount_per_player

total_bid_amount = submission['Bid Amount($)'].sum()
total_bid_amount
```

Update submission csv file

```python
# Ensure the DataFrame has the correct columns
submission = submission[['Name', 'Expected Number of Hits Next Season', 'Bid Amount($)']]

#save the result
submission.to_csv('submission.csv', index = False)
submission
```
